```python
import numpy as np
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
```

```python
def split_data(dat):
    nctot = dat.shape[1]
    nc = int(nctot/2)
    par = dat.iloc[:,0:nc]   
    perp = dat.iloc[:,nc:nctot]
    perp.reindex_like(par)

    return(par,perp)
```

```python
#prefix = '~/../../mnt/c/Users/Jcv5ff/Google Drive/UMN/Term 6 - 2021 Spring/Bio Phys/'
#folder = 'Videos Kinetics/'
#project = '03-04 - control , EutM , L , M+L kinetics/'
#sample = 'Sorted/EutM center/'
#name = 'mean_std'
prefix = '~/../../mnt/c/Users/Jcv5ff/Desktop/'
folder = 'Well/'
name = 'mean_std'

suffix = " no plasmid" 
pathin = prefix + folder + name + suffix + ".csv"
dfin = pd.read_csv( pathin )
df1,df2 = split_data(dfin)
dfmean_np = df1.iloc[:,1]
dfstd_np = df2.iloc[:,1]
# read time once
tdf = dfin.iloc[:,0]

suffix = " L center" 
pathin = prefix + folder + name + suffix + ".csv"
dfin = pd.read_csv( pathin )
df1,df2 = split_data(dfin)
dfmean_lc = df1.iloc[:,1]
dfstd_lc = df2.iloc[:,1]

suffix = " L side" 
pathin = prefix + folder + name + suffix + ".csv"
dfin = pd.read_csv( pathin )
df1,df2 = split_data(dfin)
dfmean_ls = df1.iloc[:,1]
dfstd_ls = df2.iloc[:,1]

suffix = " M center" 
pathin = prefix + folder + name + suffix + ".csv"
dfin = pd.read_csv( pathin )
df1,df2 = split_data(dfin)
dfmean_mc = df1.iloc[:,1]
dfstd_mc = df2.iloc[:,1]

suffix = " M side" 
pathin = prefix + folder + name + suffix + ".csv"
dfin = pd.read_csv( pathin )
df1,df2 = split_data(dfin)
dfmean_ms = df1.iloc[:,1]
dfstd_ms = df2.iloc[:,1]
```

```python
fig , ax = plt.subplots()
plt.errorbar(tdf,dfmean_np,yerr=dfstd_np,fmt='o',label="no plasmid")
plt.errorbar(tdf,dfmean_lc,yerr=dfstd_lc,fmt='o',label="L center")
plt.errorbar(tdf,dfmean_ls,yerr=dfstd_ls,fmt='o',label="L side")
plt.errorbar(tdf,dfmean_mc,yerr=dfstd_mc,fmt='o',label="M center")
plt.errorbar(tdf,dfmean_ms,yerr=dfstd_ms,fmt='o',label="M side")
ax.set_xlabel("Time [hr]")
ax.set_ylabel("Mean Brightness [count]")
plt.legend()
plt.show()
```

```python
fig , ax = plt.subplots()
plt.scatter(tdf,dfstd_np,label="no plasmid")
plt.scatter(tdf,dfstd_lc,label="L center")
plt.scatter(tdf,dfstd_ls,label="L side")
plt.scatter(tdf,dfstd_mc,label="M center")
plt.scatter(tdf,dfstd_ms,label="M side")
ax.set_xlabel("Time [hr]")
ax.set_ylabel("Std Dev Brightness [count]")
plt.legend()
plt.show()
```
